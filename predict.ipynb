```python
#上面我们训练了一些模型。下面我们开始使用这些模型去识别手写数字
import tensorflow as tf
import numpy as np
new_model = tf.keras.models.load_model('test_model.data')
#打印模型数据的摘要信息
new_model.summary()
#然后我们需要准备一下我们手写的数字图片(testsets/*.png)
#将这些图片灰度化，因为我们不关系这个数字是用什么颜色写的，只关心某个像素点的亮度即灰度。
#彩色每个像素是由一个长度为3的数组表示的，如[21,232,100],分别表示[R,G,B],
#处理图片我们需要用到opencv-python库。
import cv2
#如果你在处理的过程中想要输出看下图片什么样子。你还需要matplotlib库
from matplotlib import pyplot as plt

#读取图片
img = cv2.imread('testsets/num_6.png')
#因为我们的训练模型的时候使用的28x28大小的图片。所以这里定义一个函数，将被识别的图片也处理成28x28大小的灰度图片
def convert2gray(im):
    gray_img = np.zeros([28,28])
    resized_image = cv2.resize(255-im, (28, 28)) 
    for x in range(28):
        for y in range(28):
            gray_img[x][y] = sum(resized_image[x][y])//3 #取RGB像素的平均值作为灰度
    #如果你想看下处理之后的图片可以用下面注释的语句
    #plt.imshow(gray_img)
    return gray_img

gray_img = convert2gray(img)
#由于我们的数据只有一组，所以我们需要将这一个二维的图片加上索引变成三维，即相当于把图片摞起来，变成一摞，只不过这一摞只有一张图片而已
images = np.reshape(gray_img,(1,28,28))
#将0-255的像素值缩放到0-1之间
images = images / 255.0
#使用模型识别这些测试图片
predicted = new_model.predict(images)
#寻找数组中值(值越大表示匹配度越高)最大的索引，即为识别值，因为下标0-9对应着训练的时候的数字0-9
result = np.argmax(predicted)
print(result) #本示例使用的是num_6.png，代表图片上写的是6，所以这里的输出应该是6，如果不是就说明识别错误，训练的素材不够充足。一般来讲只要写的不是特别歪七八扭的识别正确率在95%以上。


#以上这些都还只是机器学习领域的helloworld.只是最最最简单的例子而已。
#总结一下机器学习其实可以抽象成函数：
# y = f(x,params)
# params = g(ey,y)
# x : 输入
# y : 实际输出
# ey : 期望输出
# d : abs(ey - y)
#通过不断的调整params来使得d最小。对应到tensorflow中,params相当于model.等训练了相当多的数据的时候，随意输入x，通过与params计算得到y。
#机器学习得到的数据不可能达到100%的正确率。只能通过大量的测试说明准确率超过99.99999...%，但是绝对不会是100%。因为训练数据是有限集合，而测试数据是无限集合。
```
