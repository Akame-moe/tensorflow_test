```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,5)
#y = 2*x+1
y = x**2
plt.plot(x,y)
#plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2
plt.figure()
plt.plot(x,y1)
plt.figure(num=2,figsize=(8,5))
plt.plot(x,y2,color='red',linewidth=1.0,linestyle='--')
plt.show()
```

```python
x = np.linspace(0,10,50)
y1 = 10*x+1
y2 = x**2
plt.figure()
plt.plot(x,y1,label='test')
plt.xlim((0,10))
plt.ylim((0,100))
plt.xlabel('XXXX')
plt.ylabel('YYYY')
new_ticks = np.linspace(0,10,11)
plt.xticks(new_ticks)
plt.yticks([60,90,100],['C','B','A'])
plt.legend()
plt.show()
```

```python
x = np.linspace(-3,3,50)
y = 2 * x + 1
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y)
#get current axis
ax = plt.gca()
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
#s:size
plt.scatter(1,3,s=100,color='blue')
plt.plot([1,1],[0,3],linestyle='--',color='red',linewidth=2.0)

#method 1
#plt.annotate(r'$2x+1=3$',xy=(1,3),xycoords='data',xytext=(+30,-30))
plt.annotate(r'$2x+1=y\ \ (x=1,y=3)$',xy=(1.3,2.5))
plt.show()
```

```python
from matplotlib import animation
fig,ax = plt.subplots()
x = np.arange(0.2*np.pi,0.01)
line, = ax.plot(x,np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x+i/10))
    return line,

def init_animate():
    line.set_ydata(np.sin(x))
    return line,
#blit:更新整个图，或者只是变化的点
ani = animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init_animate,interval=20,blit=False)
plt.show()
```
